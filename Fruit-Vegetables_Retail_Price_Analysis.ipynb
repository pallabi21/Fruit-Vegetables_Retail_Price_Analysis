```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
#dataset
fruit_prices_file = "Fruit-Prices-2022.csv"
vegetable_prices_file = "Vegetable-Prices-2022.csv"
```

```python
fruit_prices_df = pd.read_csv(fruit_prices_file)
vegetable_prices_df = pd.read_csv(vegetable_prices_file)
```

```python
# seaborn style
sns.set(style="whitegrid")
```

```python
print (fruit_prices_df)
```

```python
#top 10 most expensive fruits by retail price

top_fruits = fruit_prices_df.sort_values(by="RetailPrice", ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x="RetailPrice", y="Fruit", data=top_fruits, hue="Form")
plt.title("Top 10 Most Expensive Fruits by Retail Price")
plt.xlabel("Retail Price (per unit)")
plt.ylabel("Fruit")
plt.legend(title="Form")
plt.show()
```

```python
## Top 10 most expensive vegetables by retail price
top_vegetables = vegetable_prices_df.sort_values(by="RetailPrice", ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x="RetailPrice", y="Vegetable", data=top_vegetables, hue="Form")
plt.title("Top 10 Most Expensive Vegetables by Retail Price")
plt.xlabel("Retail Price (per unit)")
plt.ylabel("Vegetable")
plt.legend(title="Form")
plt.show()
```

```python
# Compare average retail price of fruits vs. vegetables
avg_fruit_price = fruit_prices_df["RetailPrice"].mean()
avg_veg_price = vegetable_prices_df["RetailPrice"].mean()

plt.figure(figsize=(6, 6))
sns.barplot(x=["Fruits", "Vegetables"], y=[avg_fruit_price, avg_veg_price], palette=["#FF9999", "#66B2FF"])
plt.title("Average Retail Price: Fruits vs Vegetables")
plt.ylabel("Average Price (per unit)")
plt.show()
```

```python
# Distribution of Cup Equivalent Prices
plt.figure(figsize=(12, 6))
sns.kdeplot(fruit_prices_df["CupEquivalentPrice"], label="Fruits", shade=True)
sns.kdeplot(vegetable_prices_df["CupEquivalentPrice"], label="Vegetables", shade=True)
plt.title("Distribution of Cup Equivalent Prices")
plt.xlabel("Cup Equivalent Price")
plt.legend()
plt.show()
```

```python
# Compare average retail price of fruits vs. vegetables using a pie chart
avg_prices = [fruit_prices_df["RetailPrice"].mean(), vegetable_prices_df["RetailPrice"].mean()]
labels = ["Fruits", "Vegetables"]
plt.figure(figsize=(6, 6))
plt.pie(avg_prices, labels=labels, autopct='%1.1f%%', colors=["#FF9999", "#66B2FF"], startangle=140)
plt.title("Average Retail Price Distribution: Fruits vs Vegetables")
plt.show()
```

```python
# Trend Analysis of Retail Prices using scatter plots with regression lines
plt.figure(figsize=(12, 6))
sns.regplot(data=fruit_prices_df, x=fruit_prices_df.index, y="RetailPrice", label="Fruits", scatter_kws={"color": "#FF9999"}, line_kws={"color": "red"})
sns.regplot(data=vegetable_prices_df, x=vegetable_prices_df.index, y="RetailPrice", label="Vegetables", scatter_kws={"color": "#66B2FF"}, line_kws={"color": "blue"})
plt.title("Trend Analysis of Retail Prices")
plt.xlabel("Index")
plt.ylabel("Retail Price")
plt.legend()
plt.show()
```
